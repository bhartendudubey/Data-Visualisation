```python
import pandas as pd

#loading dataset
data = pd.read_csv("D:/internity/iris.csv")
```

```python
print(data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
#Scatter Plot
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(data['sepal.length'], data['sepal.width'])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
#Line Chart
```

```python
# get columns to plot
columns = data.columns
# create x data
x_data = range(0, data.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, data[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

```python
#Histogram
```

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(data['sepal.length'])
# set title and labels
ax.set_title('sepal.width')
ax.set_xlabel('petal.length')
ax.set_ylabel('petal.width')
```

```python
# coloring each data-point by its class
```

```python
# create color dictionary
colors = {'Setosa':'r', 'Versicolor':'g', 'Virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(data['sepal.length'])):
    ax.scatter(data['sepal.length'][i], data['sepal.width'][i],color=colors[data['variety'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
#Bar Chart
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data1 = data['sepal.length'].value_counts() 
# get x and y data 
points = data1.index 
frequency = data1.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Sepal length') 
ax.set_xlabel('P') 
ax.set_ylabel('Frequency')
```
