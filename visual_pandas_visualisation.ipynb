```python
import pandas as pd

#loading dataset
data = pd.read_csv("D:/internity/iris.csv")

print(data.head())
```

```python
#Scatter Plot
data.plot.scatter(x='sepal.length', y='sepal.width', title='Iris Dataset')
```

```python
#Line Chart
data.drop(['variety'], axis=1).plot.line(title='Iris Dataset')
```

```python
#Histogram
data['petal.length'].plot.hist()
```

```python
#To create multiple histograms
data.plot.hist(subplots=True, layout=(2,2), figsize=(10, 10), bins=20)
```

```python
#Bar chart
data['sepal.width'].value_counts().sort_index().plot.bar()
```

```python
#Horizontal Bar chart
data['sepal.width'].value_counts().sort_index().plot.barh()
```
